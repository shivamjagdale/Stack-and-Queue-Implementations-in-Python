```python
# Queue Implementation using Linked List in Python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class Queue():
    def __init__(self):
        self.front = None
        self.rear = None
        
    def enqueue(self):
        x = int(input("Enter the element you want to be inserted in the stack"))
        n1 = Node(x)
        
        if self.front is None:
            self.front = n1
            self.rear = n1
            
        else:
            n2 = Node(x)
            self.rear.next = n2
            self.rear = n2
            
    def dequeue(self):
        if self.front is None:
            print("The queue is empty can't delete anything")
            
        elif self.front==self.rear:
            print("The element that has been deleted is :- ", self.front.data)
            self.front = None
        
        else:
            print("The element that has been deleted is :- ", self.front.data)
            temp = self.front
            self.front = temp.next
            temp=None
            
    def display(self):
        if self.front is None:
            print("Nothing to display cause the link is empty")
        else:
            print("The elements in the queue are :- ")
            temp = self.front 
            
            while temp:
                print(temp.data)
                temp = temp.next

s = Queue()

while(1):
    print("Enter from the menu below \n1-Enqueue Operation \n2-Dequeue Operation \n3-Display Operation \n Type anything to exit")
    
    str = input()
    
    if str=="1":
        print("Enqueue Operation")
        print("--------------------")
        s.enqueue()
        
    elif str=="2":
        print("Dequeue Operation")
        print("-------------------")
        s.dequeue()
        
    elif str=="3":
        print("Display Operation")
        print("-------------------")
        s.display()
        
    else:
        print("You have been exited from the program")
```
