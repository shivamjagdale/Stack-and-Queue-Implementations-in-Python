```python
# Stack Implementation using Linked List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class Stack:
    def __init__(self):
        self.top = None
     
    def push(self):
        x = int(input("Enter the element to be inserted into the stack"))
        
        n1 = Node(x)
        
        if self.top is None:
            self.top = n1
            self.top.next = None
        else:
            n2 = Node(x)
            n2.next = self.top
            self.top = n2
            
    def pop(self):
        
        if self.top is None:
            print("The stack is empty")
            
        elif self.top.next is None:
            print("Popped element is :-", self.top.data)
            self.top = None
            
        else:
            print("Popped element is :-", self.top.data)
            temp = self.top
            self.top = temp.next
            temp = None
            
      
    def display(self):
        if self.top is None:
            print("The stack is empty")
        else:
            print("Elements of the stack are :- ")
            temp = self.top
            
            while temp:
                print(temp.data)
                temp = temp.next
            
            print("The top of the stack is :- ", self.top.data)
            print("---------------------------------------------")
            
        
s = Stack()        
while(1):
    print("Enter from the menu below \n1-Push Operation \n2-Pop Operation \n3-Display Operation \n Type anything to exit.")
    
    str = input()
    
    if str=="1":
        print("Push Operation")
        print("---------------")
        s.push()
        
    elif str=="2":
        print("Pop Operation")
        print("---------------")
        s.pop()
        
    elif str=="3":
        print("Display Operation")
        print("---------------")
        s.display()
        
    else:
        print("You have exited from the program ")
```
